```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import io
from PIL import Image
import time
```

```python
#This is the one
wd=webdriver.Chrome()
wd.get("https://www.amazon.in/s?k=sambalpuri+saree&rh=n%3A1968256031%2Cp_n_feature_two_browse-bin%3A15650215031&dc&ds=v1%3AmloGa5Mb6HceJgytjvi00rCNWOtIRJ6yJV%2FSC06JrB0&crid=3NSIY3CZ2Q516&qid=1674905169&rnid=3959669031&sprefix=sambalpuri+saree%2Caps%2C248&ref=sr_nr_p_n_feature_two_browse-bin_12")
image_urls=[]
for _ in range(48):
	time.sleep(3)
	wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
	sr=wd.find_elements(By.CLASS_NAME,'s-image')
	for s in sr:
		image_urls.append(s.get_attribute('src'))
	nxtbtn=wd.find_element(By.CLASS_NAME,"s-pagination-next")
	nxtbtn.click()
    
def download_image(download_path, url, file_name):
	try:
		image_content = requests.get(url).content
		image_file = io.BytesIO(image_content)
		image = Image.open(image_file)
		file_path = download_path + file_name

		with open(file_path, "wb") as f:
			image.save(f, "JPEG")

		print("Success")
	except Exception as e:
		print('FAILED -', e)

#print(image_urls)
i=0
for image in image_urls:
	download_image("sambalpuri_saree_amazon/",image,str(i)+'.jpg')
	i+=1
```
